```python
## scraping data part 1 ## 


from bs4 import BeautifulSoup
import requests, bs4

```

```python

## this calss only work  with webside http://statisticalatlas.com/county/Pennsylvania/Montgomery-County/Overview
class Montgomery_County():
    """
    This class is only work with webside:
    
    http://statisticalatlas.com/county/Pennsylvania/Montgomery-County/Overview
    
    
    The main function of this calss is get all township of Montgomery County's links. 
    
        
    
    
    """

    def __init__(self,*args,link,**kwargs):
        #super(Montgomery_County,self).__init__(*args,**kwargs)
        
        if link == "http://statisticalatlas.com/county/Pennsylvania/Montgomery-County/Overview":
            self.__link = link
        else:
            raise TypeError("plz have a right web, http://statisticalatlas.com/county/Pennsylvania/Montgomery-County/Overview")
        
    @property
    def link(self):
        return self.__link
    
    
    
    def get_web_soup(self):
        """
        This function get the soup of a link
        
        return bs4.BeautifulSoup type 
        
        """
        web = requests.get(self.link)
        soup = BeautifulSoup(web.text, "html.parser")
        return soup
    
## assume    Boroughs , Townships Neighboring as twp
    def get_all_twp_link(self):
        
        """
        This function get all township's link from a link
        
        return a dic type
        """
        
        dic ={}   
        twp_web =[]
        soup = self.get_web_soup()
        for d in soup.find_all(class_="info-table-tr row"):
                for i in d.find_all(class_= "b info-table-title-td col-sm-3"):
                    if (i.get_text() == "Boroughs: " or i.get_text() == "Townships: " or i.get_text() == "Neighboring Counties: " ):
                        for n in d.find_all("a"):
                            name = n.get_text()
                            name = name.upper()
                            twp_web.append(name)
                            l = "http://statisticalatlas.com" + n.get("href")
                            dic[name] = (name,l)
        return dic
    
    
    def __str__(self):
        """
        This function shows information 
        """
        x =  self.link
        return x
    

   
    
    
    



```

```python
class Montgomery_County_Crawler(Montgomery_County):
    """
    This derived class that inherits from this parent class which is Montgomery_County class.
    
    """
    
    def __init__(self,*args,name,dic,**kwargs):
        
        super(Montgomery_County_Crawler,self).__init__(*args,**kwargs)
        
        self.__name = name
        self.__dic = dic
        
     
    
    @property
    def name(self):
        return self.__name
    
    
    @property
    def dic(self):
        return self.__dic
    
    
    
    def get_twp_link(self):
        
        if (not self.name in self.dic):
            print("Unknown twp " + self.name)
        else:
            url = self.dic[self.name][1]
            
            
        return url
        
        
    
    
    def getTwp(self):
        """ 
        Retrieve a website based on the subject short name and return it as a BeautifulSoup object. 
        Assumes an existing subjects dictionary.
        """
        if (not self.name in self.dic):
            print("Unknown twp " + self.name)
        else:
            url = self.dic[self.name][1]
            web = requests.get(url)
            web.encoding='utf8'
            nameSoup = bs4.BeautifulSoup(web.text,"html.parser")
            return nameSoup
        
        
        
         ## Get Household Income ##

    def getIncome(self):    
        soup = self.getTwp()
        #get webside of Household Income
        for row in soup.find_all(class_ = "nav nav-stacked"):
                b= row.find_all("a")
        for i in b:
            if ( i.text.strip() == "Household Income"):
                l= "http://statisticalatlas.com" + i.get("href")

        # read webside of Household Income  
        web = requests.get(l)
        web.encoding='utf8'
        income = bs4.BeautifulSoup(web.text,"html.parser")

        qq= income.find(class_ = "figure-contents")
        qq = qq.find_all("title")

        ## we choose the median as household income in this twp
        ## we can see 100% as median of income, so income is above 100%
        count =0 
        for i in qq:
            if (i.text.strip() == "100.000%"):
                break
            count = count +1

        # output income
        return qq[count-1].text.strip()   
    
    
    
    
    
        ## Get Educational Attainment##

    def getEdu(self):    
        soup = self.getTwp()
        #get webside of Educational Attainment
        for row in soup.find_all(class_ = "nav nav-stacked"):
                b= row.find_all("a")
        for i in b:
            if ( i.text.strip() == "Educational Attainment"):
                l= "http://statisticalatlas.com" + i.get("href")

        # read webside of Educational Attainment  
        web = requests.get(l)
        web.encoding='utf8'
        income = bs4.BeautifulSoup(web.text,"html.parser")

        qq= income.find(class_ = "figure-contents")
        qq = qq.find_all("text")

        ## we choose the counts of higher degree as highger educational Attainment in this twp
        ## find the number of counts 

        return qq[-5].text.strip()
    
    
    
    def getEmployment (self):
        soup = self.getTwp()
        #get webside of Educational Attainment
        for row in soup.find_all(class_ = "nav nav-stacked"):
                b= row.find_all("a")
        for i in b:
            if ( i.text.strip() == "Employment Status"):
                l= "http://statisticalatlas.com" + i.get("href")

        # read webside of Employment Status  
        web = requests.get(l)
        web.encoding='utf8'
        income = bs4.BeautifulSoup(web.text,"html.parser")

        qq= income.find(class_ = "figure-contents")
        qq = qq.find_all("text")

        ## we choose the counts of Employment  in this twp
        ## find the number of counts 

        return qq[-5].text.strip()
    
    
    
    
    
    
    
    def getRace_W (self):
        soup = self.getTwp()
        #get webside of race
        for row in soup.find_all(class_ = "nav nav-stacked"):
                b= row.find_all("a")
        for i in b:
            if ( i.text.strip() == "Race and Ethnicity"):
                l= "http://statisticalatlas.com" + i.get("href")

        # read webside of race  
        web = requests.get(l)
        web.encoding='utf8'
        race = bs4.BeautifulSoup(web.text,"html.parser")

        qq= race.find(class_ = "figure-contents")
        qq = qq.find_all(fill="#000000")

        ## we choose the counts of race of white people  in this twp
        ## find the number of counts 

        return qq[-11].text.strip()
    
    
    def Population(self):
        a = self.getTwp()
        for row in a.find_all(alt="Population",title="Population"):
            for i in row.find_all("td"):
                return i.text.strip()
            
            
    def Households(self):
        a = self.getTwp()
        for row in a.find_all(alt="Households",title="Households"):
            for i in row.find_all("td"):
                return i.text.strip()
             
        


    def __str__(self):
        print( self.name)
        print("link: "  + self.get_twp_link())
        print("Population: " + self.Population())
        print("Households: " + self.Households())
        print("Income: "+ self.getIncome())
        print("High Eduction: " + self.getEdu())
        print("Employment: " + self.getEmployment())
        print("White people: " + self.getRace_W())
        
        
    
       
    
```
